```python
"""
The distplot functions create a histogram of values of a given dataframe.
If the DataFrame contains a 'stable' column, it can plot just the stable values.
"""


def distplot_unstable(dataframe,attribute):
    values = pd.to_numeric(dataframe['stable'])
    unstable = (values==0)
    df = dataframe[unstable]
    df = pd.to_numeric(df[attribute].tolist())
    return plt.hist(df)


def distplot(dataframe,attribute):
    values = pd.to_numeric(dataframe[attribute].tolist())
    return plt.hist(values)
    
def avg_entropy(dataframe,topology):
    """
    Takes as input a dataframe that contains keys: 'sequence','dssp',and 'Si_values'.
    Returns the average entropy value for a specified dssp and the standard deviation.
    """
    specified_dssp_df = dataframe.loc[dataframe['dssp']==topology]
    dssp_entropies_str = specified_dssp_df['Si_values'].tolist()
    dssp_entropies_float = list(map(float,dssp_entropies_str))
    avg_entropy = np.mean(dssp_entropies_float)
    std_entropy = np.std(dssp_entropies_float)
    return avg_entropy, std_entropy

def residues_analysis(dataframe, topology): #topology is in the form of E,H, or L
    """
    Expedites process of graphing and anaylising residues dataframes.
    """
    if topology == 'H':
        topology_type = 'Helix'
    elif topology == 'E':
        topology_type = 'Beta Sheet'
    elif topology == 'L':
        topology_type = 'Loop'
    topology_df = dataframe.loc[dataframe['dssp']==topology]
    topology_avg, topology_std = avg_entropy(topology_df,topology)
    
    print('Avg',topology_type,'entropy is:',topology_avg)
    print(topology_type,'std is:',topology_std)
    print('N of',topology_type,'is:',len(topology_df))
    print(topology_type,'distribution:')
    print(distplot(topology_df,'Si_values'))
    
    
```

```python
def to_float(string):
    """
    Takes a string that contains the format of a list within
    and removes characters that impede the float function
    to work
    
    >>> df = pd.read_csv('/home/jupyter/tacc-work/Jan/proteins.df/rosettafinal_df.csv')
    >>> type(df['Si_values'][0])
    <class 'str'>
    
    >>> type(to_float(df['Si_values'][0]))
    <class 'list'>
    
    """
    try:
        string = string[1:-2]
        string = string.replace("'","")
        string = string.replace(",","")
        lst = string.split()
        return [float(i) for i in lst]
    except TypeError:
        print('string is:',string)
```

```python
def ks_df(df1, df2):
    """
    Creates a dataframe comparing ks values for two 
    dataframes (df1, df2).
    """
    ks_df = pd.DataFrame()

    ks_df['S_PC'] = ks_2samp(df1['S_PC'].values.tolist(),df2['S_PC'].values.tolist())

    ks_df['SumH_entropies'] = ks_2samp(df1['SumH_entropies'].values.tolist(),df2['SumH_entropies'].values.tolist())

    ks_df['SumL_entropies'] = ks_2samp(df1['SumL_entropies'].values.tolist(),df2['SumL_entropies'].values.tolist())

    ks_df['SumE_entropies'] = ks_2samp(df1['SumE_entropies'].values.tolist(),df2['SumE_entropies'].values.tolist())

    ks_df['Mean_H_entropy'] = ks_2samp(df1['Mean_H_entropy'].values.tolist(),df2['Mean_H_entropy'].values.tolist())

    ks_df['Mean_L_entropy'] = ks_2samp(df1['Mean_L_entropy'].values.tolist(),df2['Mean_L_entropy'].values.tolist())

    ks_df['Mean_E_entropy'] = ks_2samp(df1['Mean_E_entropy'].values.tolist(),df2['Mean_E_entropy'].values.tolist())

    ks_df['Mean_res_entropy'] = ks_2samp(df1['Mean_res_entropy'].values.tolist(),df2['Mean_res_entropy'].values.tolist())

    ks_df['index'] = ['KS stat','pvalue']
    ks_df.set_index('index',inplace=True)
    return ks_df

def conf_matrix_values(value_tuple):
    """
    Takes tuple returned by random_forest() and 
    returns dataframes corresponding to the 
    true positive, true negative, false 
    positive and false negative values. 
    """
    df = value_tuple[0]
    train = value_tuple[1]
    test  = value_tuple[2]
    features = value_tuple[3]
    target = value_tuple[4]
    pred_values = value_tuple[5]
    
    test['pred_values'] = pred_values

    actual_stable = test[test['stable?']==1]
    tp = actual_stable[actual_stable['pred_values']==1]
    

    fn = actual_stable[actual_stable['pred_values']==0]
    

    actual_unstable = test[test['stable?']==0]
    fp = actual_unstable[actual_unstable['pred_values']==1]
    

    tn = actual_unstable[actual_unstable['pred_values']==0]
    
    return tp, tn, fp, fn
```
