```python
print("Avg TP max H Entropy: ",np.mean(tp['H_max_entropy']))
print("Avg TN max H Entropy: ",np.mean(tn['H_max_entropy']))
print("Avg FN max H Entropy: ",np.mean(fn['H_max_entropy']))
print("Avg FP max H Entropy: ",np.mean(fp['H_max_entropy']))
```
