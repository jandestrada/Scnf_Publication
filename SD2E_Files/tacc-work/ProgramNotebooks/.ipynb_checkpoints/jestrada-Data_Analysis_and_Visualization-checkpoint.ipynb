## Main Functions

```python
def stability_probabilities_barplt(feature,step=0.1,df=open_df(),topology_label='all',stability_bool=False,stable_col='stable?',by_binary=False):
    """
    Takes a dataframe and returns the list of probabilities that the protein will be either stable 
    or unstable for each value in the bin_list.
    
    Takes a list of bin values, a target feature, a topoogy,a stability target (stable or unstable),
    and a name for the stability column to use as reference. 
    """
    #get bin_list
    bin_list = create_bin_list(df, feature, step)
    
    #create a df with binned_values
    df = get_binned_df(df, feature, bin_list=bin_list,stable_col=stable_col,by_binary=by_binary)
    
    #make a list of tuple probabilites of being stable or unstable depending on bin value
    probabilities_list = get_probabilities_list(df, bin_list, stable_col=stable_col) 

    print("PROBABILITIES LIST IS: ",probabilities_list)
    
    stability_list = specify_stability(probabilities_list, stability_bool=stability_bool)

    print("STABILITY LIST IS: ",stability_list)
    
    f,ax = bar_plot(bin_list, stability_list, stability_bool,feature, topology=topology_label)
    return f,ax
```

```python
def dor_per_bin(df,feature, step=0.1,by_binary=False,pred_col='predicted_values',topology_name='all'):
    #get bin_list
    bin_list = create_bin_list(df, feature, step)
    
    #add binned values to df
    bin_df = get_binned_df(df, feature, bin_list=bin_list, by_binary=by_binary,pred_col=pred_col)
    
    #initialize list of DOR values
    DOR_values=[]
    
    #add DOR value per bin value
    for i in bin_list:
        df = bin_df[bin_df['binned_values']==i]
        DOR_values.append(dor(df,pred_col=pred_col))
    
    #plot results
    plot_dor(bin_list=bin_list, feature=feature, topology=topology_name,dor_list=DOR_values)
    
    return DOR_values
```

```python
def roc_graphs(df, feature, pred_proba, add_pred_col=True, pred_values=None, pred_col='predicted_values', stable_val=1, stable_col='stable?',step=.1, decimal_point=2, auc_cutoff=.965, just_baseline=False, prnt=True, as_table=False,topology_label='all'):
    #prepare dataframe 
    if add_pred_col:
        df['predicted_values'] = pred_values
    probabilities = get_predicted_probabilities(pred_proba)
    
    df['predicted_probabilities'] = probabilities
    
    bin_list= create_bin_list(df, feature, step)

    #add binned_values to df
    bin_df = get_binned_df(df, feature, bin_list=bin_list, pred_col=pred_col).reset_index(drop=True)
    
    #separate into stable and unstable
    stable = df[df[stable_col]==stable_val]
    unstable = df[df[stable_col]!=stable_val]
    
    #create lists for cutoffs
    stable_range, unstable_range = create_cutoff_lists(stable, feature, step=step, decimal_point=decimal_point),\
    create_cutoff_lists(unstable, feature, step=step, decimal_point=decimal_point)
    
    min_values = unstable_range
    max_values = stable_range
    if prnt:
        print('min values: ',min_values)
        print('max values: ',max_values)
    
    #show table
    if as_table:
        auc_df = auc_table(bin_df, min_values, max_values, stable_col, feature, topology=topology_label)
        return auc_df
    
    #plot curves
    plot_roc_curves(min_values, max_values, bin_df, stable_col, feature, auc_cutoff, just_baseline=just_baseline)
    
```

```python
def chk_prob(entropy_val, feature, topology, stable_col='stable?', df=open_df(),step=0.1,label='stable'):#TODO: add a test so entropy_val can be turned into a list if it is not
    """
    Entropy val must be inputted as list-like.
    """
    
    bin_list = create_bin_list(df, feature, step)
    
    ind = digitize_bins(df, feature ,bin_list )
    binned_values = []
    for i in ind.tolist():
        binned_values.append(bin_list[i])

    df['binned_values'] = binned_values
    df_foo = df[['binned_values',stable_col]]
    print(df_foo.binned_values.value_counts())
    for value in entropy_val:
        ent_df = df[df['binned_values']==value]
        n_stable,n_unstable = get_n_stable(ent_df)
        result=-1
        print('length of values: ',len(ent_df))
        print('number of unstable:', n_unstable) 
        print('number of stable: ',n_stable)
        if label=='stable':
            result= n_stable/(n_unstable+n_stable+.000001)
        elif label=='unstable':
            result = n_unstable/(n_stable+n_unstable+.000001)
        label = label
        print("\nProbability %s value is %s == %s \n" %(value, label, result))
```

```python
def avg_min_max(df_list, topology, feature, stable_col='stable?',stable_val=1,prnt=True):
    min_list_stable = []
    max_list_stable = []
    min_list_unstable = []
    max_list_unstable = []
    
    for df in df_list:
        min_stable, max_stable, min_unstable, max_unstable = get_min_max(df, topology, feature, stable_col,stable_val,prnt=False)
        min_list_stable.append(min_stable)
        max_list_stable.append(max_stable)
        min_list_unstable.append(min_unstable)
        max_list_unstable.append(max_unstable)
    
    avg_min_stable = np.mean(min_list_stable)
    avg_max_stable = np.mean(max_list_stable)
    avg_min_unstable = np.mean(min_list_unstable)
    avg_max_unstable = np.mean(max_list_stable)
    
    if prnt:
        print("\nAVG MIN for STABLE == ", avg_min_stable)
        print("AVG MAX for STABLE == ",avg_max_stable)
        print("AVG MIN for UNSTABLE ==", avg_min_unstable)
        print("AVG MAX for UNSTABLE ==",avg_max_unstable)
    
    return avg_min_stable,avg_max_stable,avg_min_unstable,avg_max_unstable
```

### `stability_probabilities_barplt` helper functions

```python
def create_bin_list(df, feature, step):
    #create bin_list
    max_feature_val = round(max(df[feature]),2)
    min_feature_val = round(min(df[feature]),2)
    bin_list = np.arange(min_feature_val,max_feature_val+step,step)
    return bin_list
```

```python
def get_binned_df(df,feature, bin_list,pred_col=None, stable_col='stable?',by_binary=False):
    
    # get indices through digitize_bins
    ind = digitize_bins(df=df, feature=feature, bin_values_list=bin_list)
    
    # get binned_values
    binned_values = get_binned_values(ind, bin_list=bin_list)
    
    # add binned_values to df
    df['binned_values'] = binned_values
    
    return df
```

```python
def get_probabilities_list(df, bin_list, stable_col='stable?'):
    
    probabilities_list = []
    
    for i in bin_list:
        # find where df has 
        specified_df = df[df['binned_values']==i]
        
        # find positive and negatives values to for this VAL specific df
        stable, unstable = specified_df[specified_df['stable?']==1], specified_df[specified_df[stable_col]==0]
        
        # get number of stable and unstable examples
        n_stable = len(stable)
        n_unstable = len(unstable)
        
        # get probability of positive and negative
        p_stable = n_stable/(n_stable+n_unstable+.00000000001)
        p_unstable = n_unstable/(n_stable+n_unstable+.00000000001)
        
        probabilities_list.append((p_stable,p_unstable))
    
    return probabilities_list
     
```

```python
def specify_stability(probabilities_list, stability_bool):
    stability_list = []
    tuple_inx = None
    if stability_bool:
        tuple_inx=0
    else:
        tuple_inx=1
    for i in probabilities_list:
        stability_list.append(i[tuple_inx])
    return stability_list
        
```

```python
def bar_plot(bin_list, stability_list, stability_bool, feature,topology='all'):
    """
    Function taken from https://pythonspot.com/matplotlib-bar-chart/
    """
    
    bin_list = list(map(lambda i: round(i,1), bin_list))
    objects= tuple(bin_list)
    y_pos = np.arange(len(objects))
    performance = stability_list
    label= get_label(stability_bool)
    plt.bar(y_pos, performance, align='center', alpha=0.5)
    plt.xticks(y_pos, objects)

    
    plt.xlabel('%s' %feature)
    plt.ylabel('Probability of %s' %label)
    plt.title('%s for %s topologies' %(feature, topology))
    fig, ax = plt.gcf(), plt.gca()
    return fig,ax
```

```python
def dor(df, pred_col='predicted_values',stable_col='stable?'):
    tp,tn,fp,fn = TPTN_classification(df,pred_col=pred_col,stable_col=stable_col)
    tp_count = float(len(tp))
    tn_count = float(len(tn))
    fp_count = float(len(fp))
    fn_count = float(len(fn))
    alpha = 10**-20
    DOR = ((tp_count+alpha)/(fp_count+alpha))/((fn_count+alpha)/(tn_count+alpha))
    return DOR
```

```python
def plot_dor(bin_list, dor_list, feature, topology):
    plt.bar(x=bin_list, height=dor_list, width=.1, align='center')
    plt.ylim(ymax=1000)
    plt.xlabel(feature)
    plt.ylabel('Diagnostic Odds Ratio')
    plt.title('Diagnostic Odds Ratio for %s in %s proteins' %(feature, topology))
    plt.show()
             
```

```python
def create_cutoff_lists(df, feature,step=.1, decimal_point=2):
    max_val, min_val = max(df[feature]), min(df[feature])
    range_list = list(np.arange(min_val, max_val, step=step))
    range_list = round_floats(range_list, decimal_point)
    return range_list
```

```python
def auc_table(bin_df, min_values, max_values, stable_col, feature, topology):
    """
    Returns a dataframe of AUC values.
    """
    data = []
    count=0
    for pair in product(min_values, max_values): #from itertools 
        try:
            #get cutoffs for graph
            min_cutoff, max_cutoff = pair

            #test to make sure min < max
            min_cutoff, max_cutoff = chk_min_max(min_cutoff, max_cutoff)

            #change df so you only look at values within the cutoff
            above_min = bin_df[bin_df['binned_values'] >= min_cutoff]
            temp_df = above_min[above_min['binned_values']<= max_cutoff]


            #plot the roc curve for this df
            fpr, tpr, thresholds = roc_curve(np.array(temp_df[stable_col]), np.array(temp_df['predicted_probabilities']))
            roc_auc = auc(fpr,tpr)

            #create data to be added to table
            data_instance = [feature, topology,roc_auc,min_cutoff,max_cutoff]
            data.append(data_instance)
        except ValueError:
            count+=1
            #print("ValueError with cutoff params: %0.2f, %0.2f" %(min_cutoff, max_cutoff))
            pass
    
    table = pd.DataFrame.from_records(data, columns=['FEATURE','TOPOLOGY','AUC SCORE','MIN CUTOFF','MAX CUTOFF'])
    return table
        
        
```

```python
def plot_roc_curves(min_values, max_values, bin_df, stable_col, feature, auc_cutoff, just_baseline=False):
    plt.figure() #create the figure where all the plots will be on
    
    if just_baseline:
        fpr, tpr, thresholds = roc_curve(bin_df[stable_col], bin_df['predicted_probabilities'].values)
        roc_auc = auc(fpr,tpr)
        lw = 2 #linewidth
        plt.plot(fpr, tpr,
                 lw=lw, label='ROC curve (area = %0.2f, cutoffs = %0.2f, %0.2f)' % (roc_auc, min(min_values), max(max_values)))
        plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.suptitle('Receiver operating characteristic')
        plt.title(str(feature))
        plt.legend(loc='right')
    else:
        count=0
        for pair in product(min_values, max_values): #from itertools 
            try:
                #get cutoffs for graph
                min_cutoff, max_cutoff = pair

                #test to make sure min < max
                min_cutoff, max_cutoff = chk_min_max(min_cutoff, max_cutoff)

                #change df so you only look at values within the cutoff
                above_min = bin_df[bin_df['binned_values'] >= min_cutoff]
                temp_df = above_min[above_min['binned_values']<= max_cutoff]

                #plot the roc curve for this df
                fpr, tpr, thresholds = roc_curve(temp_df[stable_col], temp_df['predicted_probabilities'].values)
                roc_auc = auc(fpr,tpr)
                if roc_auc >=auc_cutoff:
                    lw = 2 #linewidth
                    plt.plot(fpr, tpr,
                             lw=lw, label='ROC curve (area = %0.2f, cutoffs = %0.2f, %0.2f)' % (roc_auc, min_cutoff, max_cutoff))
                    plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
                    plt.xlim([0.0, 1.0])
                    plt.ylim([0.0, 1.05])
                    plt.xlabel('False Positive Rate')
                    plt.ylabel('True Positive Rate')
                    plt.suptitle('Receiver operating characteristic')
                    plt.title(str(feature))
                    plt.legend(loc='right')
            except ValueError:
                count+=1
                #print("ValueError with cutoff params: %0.2f, %0.2f" %(min_cutoff, max_cutoff))
                pass
        print('Skipped',count,'entries')
    plt.show()
```

### `chk_probs` helper functions

```python
def digitize_bins(df,feature, bin_values_list):
    """
    Takes a dataframe, specified features, and bin values
    and returns indices for the df values for the bin values.
    
    Example usage:
    >>> df = pd.DataFrame([[0,4,2,6,3,2],[1,4,2,1,3,2],[12,4,1,2,2,4]])
    >>> digitize_bins(df, feature=0, bin_values_list=[2,4])
    array([0, 0, 2])
    
    """
    x = df[feature].values
    bins = np.array(bin_values_list)
    inds = np.digitize(x,bins,right=True)
    inds = np.array(chk_inds(inds, len(bins)))
    return inds
    
```

```python
def get_n_stable(df, stable_col='stable?', stable_value=1):
    """
    Returns a tuple of number of stable and unstable values
    """
    n_stable = len(df[df[stable_col]==stable_value])
    n_unstable = len(df[df[stable_col]!=stable_value])
    
    return n_stable,n_unstable
```

## `avg_min_max` helper functions

```python
def get_min_max(df, feature, stable_col='stable?',stable_val=1,prnt=True):
    df = df[[feature,stable_col]]
    df_stable = df[df[stable_col]==1]
    df_unstable = df[df[stable_col]!=1]
    
    min_stable = min(df_stable[feature].values)
    max_stable = max(df_stable[feature].values)
    min_unstable = min(df_unstable[feature].values)
    max_unstable = max(df_unstable[feature].values)
    
    if prnt:
        print("MIN for STABLE ==",min_stable)
        print("MAX for STABLE ==",max_stable)
        print("MIN for UNSTABLE ==",min_unstable)
        print("MAX for UNSTABLE ==",max_unstable)
    
    return min_stable,max_stable,min_unstable,max_unstable
```

### Misc Helper functions

```python
def add_topologies(df, topologies_path='/home/jupyter/tacc-work/Jan/proteins.df/topologies_df.csv'):
    """
    Adds the topologies column to a dataframe that contains a 'sequence' column.
    Need to specify the dataframe to which topologies will be added, and a file path to a csv
    file containing topologies information and a sequence conlumn.
    
    Returns the new dataframe.
    """
    #open Topologies DF
    top_df = pd.read_csv(topologies_path)
    top_df.set_index('sequence',inplace=True)

    #prepare target dataframe
    target_df = df
    target_df.set_index('sequence',inplace=True)

    #merge dataframes
    df = pd.merge(target_df,top_df,left_index=True,right_index=True)
    df = df.reset_index()

    return df

```

```python
def add_entropy_features(df, name_col='name'):
    df.drop_duplicates(inplace=True)
    df.rename(columns={name_col:'protein'},inplace=True)
    
    rosetta_df = open_df()
    
    mrgd_df = df.merge(rosetta_df[['S_PC','sequence','Mean_H_entropy','Mean_L_entropy','Mean_E_entropy','Mean_res_entropy','SumH_entropies','SumL_entropies','SumE_entropies','H_max_entropy',"L_max_entropy","E_max_entropy",'H_min_entropy',"L_min_entropy","E_min_entropy",'H_range_entropy',"L_range_entropy","E_range_entropy","stabilityscore"]], on='sequence')
    
    mrgd_df.dropna(inplace=True)
    
    return mrgd_df
```

```python
def chk_inds(ndarray, max_val):
    return [(max_val-1) if x==max_val else x for x in ndarray]
```

```python
def chk_min_max(min_cutoff, max_cutoff):
    if min_cutoff > max_cutoff:
        new_min = max_cutoff
        new_max = min_cutoff
    else:
        new_min = min_cutoff
        new_max = max_cutoff
    return new_min, new_max
```

```python
def round_floats(float_list, decimal_point=2):
    new_list = []
    for i in float_list:
        x = round(i, decimal_point)
        new_list.append(x)
    return new_list
```

```python
def spellcheck_for_protein(df,feature, min_cutoff=-.233, max_cutoff=.343, stable_col='stable?', stable_val=1):
    """
    Takes a dataframe that has already been specified for a topology as i
    """
    new_df = df
    new_df = new_df.reset_index()
    print("Length before: ",len(new_df))
    for i in range(len(new_df[feature])):
        if not min_cutoff <= new_df[feature][i] <= max_cutoff:
            print(i)
            new_df = new_df.drop(i,axis=0)
    print('Length after: ',len(new_df))
    return new_df
```

```python
def TPTN_classification(df,stable_col='stable?',pred_col='predicted_values',stable_value=1):
    actual_stable = df[df[stable_col]==stable_value]
    actual_unstable = df[df[stable_col]!=stable_value]
    
    tp = actual_stable[actual_stable[pred_col]==stable_value]
    tn = actual_unstable[actual_unstable[pred_col]!=stable_value]
    fp = actual_unstable[actual_unstable[pred_col]==stable_value]
    fn =  actual_stable[actual_stable[pred_col]!=stable_value]
    
    return tp,tn,fp,fn
```
