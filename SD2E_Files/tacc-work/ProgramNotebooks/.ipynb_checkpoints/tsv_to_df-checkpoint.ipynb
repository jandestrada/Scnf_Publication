## .tsv to dictionary
* will work with different *manager* function (`tsv_to_df`)
* `tsv_to_df` will feed this function tsv files and send the dictionaries to another function
* `tsv_to_df` will feed another function the dictionaries and store the dataframes in another directory

* FIXME: Dictionary has the header. Which is not good. Fix it.

```python
def tsv_to_dict(tsv_file): #takes as input a tsv file and returns a dictionary
    #initializes empty dictionary FORMAT: {name:{S_PC:value,reliability_n:value,total_score:value,stability_score:value}}
    S_PC_dict = {}
    
    #opens tsv file
    reader=tsv.TsvReader(open(tsv_file))
    
    #iterates through lines
    for line in reader:
        line = list(line)
        #defines variables for each value 
        filename = line[0]
        library = line[1]
        S_PC = line[2]
        reliability_n = line[2]
        Si_values = (line[3:])
        
        #update dictionary
        S_PC_dict[filename]={'S_PC':S_PC,'reliability':reliability_n,'Si_values':Si_values}
    #returns dictionary
    return S_PC_dict
```

```python
test=tsv_to_dict('Test.tsv')

df=pd.DataFrame(test)
df.drop('Si_values',axis=0) #df[['reliability','S_PC']]
df.drop('filename',axis=1)
df
```

## dictionary to S_PC_df
* will receive dictionaries from a different function, convert to dataframe using pandas, and store in a specified directory
* FIXME I am being sloppy by taking out the header from the tsv here instead of in the `tsv_to_dict` function

```python
def dict_to_SPC_df(dictionary):
    #creates a dataframe from dictionary
    df = pd.DataFrame(dictionary)
    #drops Si_values and header from tsv
    df = df.drop('Si_values',axis=0)
    df = df.drop('filename',axis=1)
    #returns dataframe
    return df
```

## Manager function
* loops over tsv file directory, running them through the functions and storing them in a target directory

```python
def tsv_to_df(tsv_dir,library_name): #takes a tsv directory, a target directory, and a library_name as inputs
    #make sure you are at file_dir 
    path_name=''
    if os.getcwd != '/home/jupyter/tacc-work/Jan/proteins.tsv/':
        os.chdir('/home/jupyter/tacc-work/Jan/proteins.tsv/')
        path_name = os.path.join(os.getcwd(),tsv_dir)
    else:
        path_name = os.path.join(os.getcwd(),tsv_dir)
    
    #iterature through tsv_dir
    tsv_list = os.listdir(tsv_dir)
     
    #iterates through tsv
    for file in tsv_list:
        #runs each file through tsv_to_dict
        dictionary = tsv_to_dict(file)
        #runs each result through dict_to_SPC_df
        df = dict_to_SPC_df(dictionary)
        #stores each result in target_dir
        df.to_csv((str(library_name)+'.df'),sep='\t')
```
