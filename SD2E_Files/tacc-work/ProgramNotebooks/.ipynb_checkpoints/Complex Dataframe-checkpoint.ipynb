```python

def max_entropies(binary_series,Si_series):
    values = []
    for i in range(len(binary_series)): #iterates through series
        try:
            values.append(max(binary_series[i]*(Si_series[i]))) #appends the multiplication of the binary_series to the Si_values
        except ValueError:
            print('Protein with index: ',i,'was not appended.')
    return values
```

```python
def dssp_to_bin(string, dssp):
    binary_list = []
    for i in string:
        if i == dssp:
            binary_list.append(1)
        else:
            binary_list.append(0)
    return binary_list

def sum_entropies(binary_series,Si_series):
    values = []
    for i in range(len(binary_series)): #iterates through series
        try:
            values.append(sum(binary_series[i]*(Si_series[i]))) #appends the multiplication of the binary_series to the Si_values
        except ValueError:
            print('Protein with index: ',i,'was not appended.')
    return values


def column_dict(df, dssp): 
    """
    Takes a merged dataframe and dssp value as inputs
    and returns a dictionary with 'sum_<dssp>_entropy as a key
    and a list 
    """
    #creates empty final dictionary
    final_dict = {}
    #creates empty values dictionary
    values_dict = {}
    #defines final key for dictionary
    final_key = 'sum_'+dssp+'_entropy'
    #iterates through dataframe protein names
    for protein_name in df.protein.values:
        #obtains a key and value from sum_entropies function
        key, value = sum_entropies(df,protein_name,dssp)
        #checks that the entry does not already exist in dictionary
        if key in values_dict:
            continue
        #creates entry if it did not exist
        values_dict[key]=value
    #defines final_dictionary with values_dict and final_key
    final_dict[final_key]=values_dict
    #returns final_dictionary
    return final_dict

def to_float(string):
    try:
        string = string[1:-2]
        string = string.replace("'","")
        string = string.replace(",","")
        lst = string.split()
        return [float(i) for i in lst]
    except TypeError:
        print('string is:',string)
```
