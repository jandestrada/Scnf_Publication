```python
import pandas as pd
import os
import sklearn.preprocessing as pp
```

```python
final_df = pd.read_csv('/home/jupyter/tacc-work/Jan/proteins.df/rosettafinal_df.csv')
final_df = final_df.set_index('sequence')
final_df = final_df.drop('Unnamed: 0',axis=1)
final_df = final_df.drop_duplicates(keep='first')
final_df = final_df.drop(final_df[final_df.index.duplicated(keep='first')].index.values)

E1T_df = pd.read_csv('/home/jupyter/sd2e-community/protein-design/data_v1_April_2018/Eva1_Dec_2017/topology.csv')
E1T_df.topology.value_counts()
E1T_df['protein'] = E1T_df['pdb_path'].map(os.path.basename)
E1T_df = E1T_df.set_index('sequence')

E2T_df = pd.read_csv('/home/jupyter/sd2e-community/protein-design/data_v1_April_2018/Eva2_Dec_2017/topology.csv')
E2T_df.topology.value_counts()
E2T_df['protein'] = E2T_df['pdb_path'].map(os.path.basename)
E2T_df = E2T_df.set_index('sequence')
E2T_df

IT_df = pd.read_csv('/home/jupyter/sd2e-community/protein-design/data_v1_April_2018/Inna_April_2016/topology.csv')
IT_df.topology.value_counts()
IT_df['protein'] = IT_df['pdb_path'].map(os.path.basename)
IT_df = IT_df.set_index('sequence')
IT_df

LT_df = pd.read_csv('/home/jupyter/sd2e-community/protein-design/data_v1_April_2018/Longxing_Dec_2017/topology.csv')
LT_df.topology.value_counts()
LT_df['protein'] = LT_df['pdb_path'].map(os.path.basename)
LT_df = LT_df.set_index('sequence')
LT_df

RT_df = pd.read_csv('/home/jupyter/sd2e-community/protein-design/data_v1_April_2018/Rocklin_2017_Science/topology.csv')
RT_df.topology.value_counts()
RT_df['protein'] = RT_df['pdb_path'].map(os.path.basename)
RT_df = RT_df.set_index('sequence')
```

```python
enc = pp.LabelEncoder()
features = df['topology'].values
new_features = enc.fit_transform(features)
print(new_features)
new_features = new_features.reshape(-1,1)
ohe = pp.OneHotEncoder(sparse=False)
final_features = ohe.fit_transform(new_features)
```

```python
df['ohe_topology'] = list(final_features)
```
