## Helper Functions

```python
def add_entropy_features(df, name_col='name'):
    df.rename(columns={name_col:'protein'},inplace=True)
    df['protein'] = df['protein'].map(canonicalize_name)
    
    rosetta_df = pd.read_csv('/home/jupyter/tacc-work/Jan/proteins.df/rosettafinal_df.csv')
    rosetta_df.drop('Unnamed: 0',axis=1,inplace=True)
    rosetta_df.rename(columns={'protein_x':'protein'},inplace=True)
    rosetta_df['protein'] = rosetta_df['protein'].map(canonicalize_name)
    
    mrgd_df = df.merge(rosetta_df[
        ['protein','S_PC','Mean_H_entropy','Mean_L_entropy','Mean_E_entropy','Mean_res_entropy','SumH_entropies','SumL_entropies','SumE_entropies']
    ], on='protein')
    
    mrgd_df.dropna(inplace=True)
    
    return mrgd_df
```

```python
def add_topologies(df):
    """
    Adds the topologies column to a dataframe that contains a 'sequence' column.
    
    Returns the new dataframe.
    """
    #open Topologies DF
    top_df = pd.read_csv('/home/jupyter/tacc-work/Jan/proteins.df/topologies_df.csv')
    top_df.set_index('sequence',inplace=True)

    #prepare target dataframe
    target_df = df
    target_df.set_index('sequence',inplace=True)

    #merge dataframes
    df = pd.merge(target_df,top_df,left_index=True,right_index=True)
    df = df.reset_index()

    return df


    
```

```python
def change_topology_x(df, topology):
    #specify topology to be looked at for main dataframe
    try:
        df = df[df['topology']==topology]
    except:
        print("Error occurred: Topology not found")
    return df
```

```python
def histogram(dataframe,attribute,labels,density=False): #takes a max of two dataframes
    if type(dataframe) == list:
        df1 = dataframe[0]
        df2 = dataframe[1]
        label1 = labels[0]
        label2 = labels[1]
        value = pd.to_numeric(df1[attribute].tolist())
        plt.hist(value,density=density,label=label1)
        value2  = pd.to_numeric(df2[attribute].tolist())
        plt.hist(value2, alpha=.75,density=density,label=label2)
        plt.legend()
        plt.title(attribute+" ("+label1+' , '+label2+')')
        plt.show()
    else:
        values = pd.to_numeric(dataframe[attribute].tolist())
        plt.hist(values, density=density, label=labels)
        plt.legend()
        plt.title(attribute+labels)
        plt.show()
```

```python
def ks_histograms(df, cutoff,target_col='mr_class_predictions',density=True):
    import math
    """
    Plots histograms of a given 
    KS value dataframe where the 
    KS values are above a given 
    cutoff
    """
    #create True Positive and False Negative values
    actual_positive = df[df['stable?']==1]
    tp = ("TP",actual_positive[actual_positive[target_col]==1.0])
    fn = ("FN",actual_positive[actual_positive[target_col]==0.0])

    #create True Negative and False Positive values
    actual_negative = df[df['stable?']==0]
    tn = ("TN",actual_negative[actual_negative[target_col]==0.0])
    fp = ("FP",actual_negative[actual_negative[target_col]==1.0])
    
    
    for pair in combinations([tp,fn,tn,fp],2):
        df1 = pair[0][1]
        label1 = pair[0][0]
        df2 = pair[1][1]
        label2 = pair[1][0]
        #print(pair[0][1],pair[1][1])
        ks_values = ks_df(df1,df2)
        signf_ks_values = ks_values[ks_values>cutoff]
        for column in signf_ks_values:
            if math.isnan(signf_ks_values[column][0]):
                continue
            histogram([df1,df2],column, labels=[label1,label2],density=density)
```

```python
def ks_stat_df(df,target_col,\
               list_of_columns=['S_PC','Mean_H_entropy','Mean_L_entropy','Mean_E_entropy',\
                                'Mean_res_entropy','SumH_entropies','SumL_entropies','SumE_entropies'],\
               cutoff=0): 
    from itertools import combinations
    
    """
    Takes a dataframe that contains y_pred and
    y_actual values and creates a dataframe
    with its ks stat values and pvalues.
    
    Must specify what the target_col is called (i.e,
    'pred_values', 'mr_class_predictions', etc)
    """
    #create True Positive and False Negative values
    actual_positive = df[df['stable?']==1]
    tp = ("TP",actual_positive[actual_positive[target_col]==1.0])
    fn = ("FN",actual_positive[actual_positive[target_col]==0.0])

    #create True Negative and False Positive values
    actual_negative = df[df['stable?']==0]
    tn = ("TN",actual_negative[actual_negative[target_col]==0.0])
    fp = ("FP",actual_negative[actual_negative[target_col]==1.0])
    
    df_final = pd.DataFrame()
    
    for pair in combinations([tp,fn,tn,fp],2):
        df1 = pair[0][1]
        df2 = pair[1][1]
        df = ks_df(df1,df2)
        df['name'] = [pair[0][0]+pair[1][0]]*len(df)
        df_final = df_final.append(df)
    
    df_final = df_final.set_index([df_final.index,df_final['name']])
    df_final.drop('name',axis=1,inplace=True)
    df_final = df_final[df_final.loc['KS stat'] > cutoff]
    return df_final
    
```

### Ben's function
Ben wrote a function that separated proteins into topologies strictly based on DSSP using the 'L' as separatators. This led our data to be categorized differently to the standard way most were separating their data in SD2E.

```python
def dssp_to_topo(dssp):
    topology = ''
    stuff = dssp.split('L')
    for letter in stuff:
        if 'E' in letter:
            topology = topology + 'E'
        elif 'H' in letter:
            topology = topology + 'H'
        else:
            pass
    return topology
```
