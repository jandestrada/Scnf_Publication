### Random Forest

```python
def random_forest(datasample='stratify', index=rosetta_index,prnt=True, topology='all', cw=None, custom_df=None, default_df=True):
    """
    Creates a random forest using a specified index and method for dealing with class imbalance. 
    Returns a the df, train, test, features, target, pred_values, and classifier varible used.
    
    >>> df, train,test,features,target, pred_values, classifier = random_forest(prnt=False)
    >>> type(pred_values)
    <class 'numpy.ndarray'>
    
    """
    
    inx_bool = (len(index)>1)
    
    #open main dataframe
    df = open_df(default=default_df,df=custom_df)

    #specify topology to be looked at for main dataframe
    if topology != 'all':
        df = change_topology(df, topology)

    #read and fix the training and testing dataframes
    train, test = train_test_df(df, datasample)          

    
  
    #change topologies if necessary
    if datasample=='downsample_A' and topology != 'all':
        try:
            train_df = change_topology(train, topology)
            test_df = change_topology(test, topology)
        except ValueError: #this was added because in any method other than downsample_A, 
            pass           #change_topology will be unnecessary since the dataframes will already have the specified columns and the others will be dropped.
    
    #check for NaN values in train and test data
    train = train.fillna(0)
    test = test.fillna(0)

    #prepare and fit predictions
    target, features, classifier = prep_run_fit(cw=cw, df=df, train=train, test=test, index=index, target_col='stable?', multifeatures=inx_bool)
    
    #TODO Write test that confirms there is more than one class in pred_values
    
    #store/print results
    pred_value, pred_proba, oob_score, acc_score, \
    class_report, conf_matrix, ras, mcc, importances, cvs = store_results(df, classifier, test, features,multifeatures=inx_bool)

    #get feature importances
    feature_importances(classifier, importances, train, features, prnt)
    
    #print results:
    print_results(prnt, pred_value, pred_proba, oob_score, acc_score, \
    class_report, conf_matrix, ras, mcc, cvs)

    return df, train, test, features, target, pred_value, classifier, pred_proba
```

```python
def open_df(default=True, df=None):
    """
    Opens the RosettaFinal dataframe stored in proteins.df library.
    """
    if default:
        df = pd.read_csv('/home/jupyter/tacc-work/Jan/proteins.df/rosettafinal_df.csv') 
        df = df.drop('Unnamed: 0',axis=1) #------------------------------\
        df['Si_values'] = df['Si_values'].map(to_float) #-----------------> Fixes to main dataframe
        df['ohe_topology'] = df['ohe_topology'].map(to_float) #----------/
        return df
    return df
```

```python
def change_topology(df, topology, by_binary=True):  #TODO: CHANGE IT SO THAT IT WOKRS WITH STRING DUDE CMON
    #specify topology to be looked at for main dataframe
    try:
        if by_binary:
            df_topology = df[df[topology]==1]
            df = clean_columns(df_topology, ['4h','HEEH','ferredoxin','HHH','EEHEE','EHEE','beta_grasp','thio','fold2', 'coil','fold4'])
        else:
            df = df[df['topology']==topology]
    except:
        print("Error occurred: Topology not found")
    
    return df
```

```python
def train_test_df(df, datasample): #TODO make the downsample_A thing better
    """
    Splits the dataframe into training and testing data based on 
    the specified datasample. 
    
    """
    if datasample=='downsample_A':
        train, test = pd.read_csv('/home/jupyter/tacc-work/Jan/proteins.df/rosetta_train.csv'), pd.read_csv('/home/jupyter/tacc-work/Jan/proteins.df/rosetta_test.csv')
        train = train.rename(columns={'Unnamed: 0':'protein'}) #----------------------\
        train['Si_values'] = train['Si_values'].map(to_float) #----------------------- \
        train['ohe_topology'] = train['ohe_topology'].map(to_float) #-------------------> Fixes to training and testing dataframes
        train = train.set_index('protein') #-------------------------------------------/
        train.fillna(value=0,inplace=True)    #------------------------------------------/ >> Fills NaN values for beta sheet entropies with 0
        test = test.rename(columns={'Unnamed: 0':'protein'}) #---------------------/
        test = test.set_index('protein') #---------------------------------------/
        test['Si_values'] = test['Si_values'].map(to_float) #------------------/
        test['ohe_topology'] = test['ohe_topology'].map(to_float) #-----------/
        test.fillna(value=0, inplace=True)
    elif datasample=='downsample_B':
        df = sample(df, direction='down')
        train, test = split_df(df)
    elif datasample=='upsample':
        df = sample(df, direction='up')
        train, test = split_df(df)
    elif datasample=='unbalanced':
        train,test = split_df(df)
    elif datasample=='stratify':
        library_list = list(np.unique(df['library'].values))
        train,test = split_df(df)
        train = stratify(train, library_list)
        test = stratify(test, library_list)
        
    else:
        print("INVALID DATASAMPLE METHOD. \n ACCEPTED VALUES ARE: 'downsample_A', 'downsample_B', 'upsample', and 'unbalanced'")
    return train, test
```

```python
def prep_run_fit(cw, df, train, test, index, target_col, multifeatures=True, clf='RandomForest',normalize=False):
    """
    Sets the target for the classifier and fits it. 
    """
    target = train[target_col]
    features = df.columns[index]
    train = train.fillna(value=0)
    test = test.fillna(value=0) 
    
    if clf=='RandomForest':
        classifier = RandomForestClassifier(n_estimators=50, n_jobs=1,random_state=0,oob_score=True,class_weight=cw)
    elif clf=='LinearRegression':
        classifier = LinearRegression(normalize=normalize)
    elif clf=='LogisticRegression':
        classifier=LogisticRegression(class_weight=cw)
    
    if multifeatures:
        rf_values = train[features]
    else:
        rf_values = train[features].values.reshape(-1,1)
    #run predictions
    classifier.fit(rf_values,target)  #train classifier on features
    return target, features, classifier
```

```python
def store_results(df, classifier, test, features, multifeatures=True,clf='RandomForest'):
    if multifeatures:
        rf_values = test[features]
    else:
        rf_values = test[features].reshape(-1,1)
    pred_value = classifier.predict(rf_values) #save predicted values to a variable
    pred_proba = classifier.predict_proba(rf_values)
    oob_score = classifier.oob_score_ #save oob score to a variable
    acc_score = accuracy_score(test['stable?'],classifier.predict(rf_values))
    class_report = classification_report(test['stable?'],classifier.predict(rf_values))
    conf_matrix= confusion_matrix(test['stable?'],classifier.predict(rf_values)) #save confusion matrix
    ras = roc_auc_score(test['stable?'],classifier.predict(rf_values)) #save roc_auc_score to variable
    mcc = matthews_corrcoef(test['stable?'],classifier.predict(rf_values)) #save matthew correlation coef to variable
    importances = classifier.feature_importances_
    cvs=np.mean(cross_val_score(classifier,rf_values.values,test['stable?'].values))
    return pred_value, pred_proba, oob_score, acc_score, class_report, conf_matrix, ras, mcc, importances, cvs
    
```

```python
def feature_importances(classifier, importances, train, features, prnt):
    """
    The following code is taken from http://scikit-learn.org/stable/auto_examples/ensemble/plot_forest_importances.html
    """
    std = np.std([tree.feature_importances_ for tree in classifier.estimators_], axis=0)
    indices = np.argsort(importances)[::-1]

    # Print the feature ranking
    if prnt==True:
        print("Feature ranking:")

        for f in range(train[features].shape[1]):
            print("%d.  %s (%f)" % (f + 1, features[indices[f]], importances[indices[f]]))

        # Plot the feature importances of the forest
        plt.figure()
        plt.title("Feature importances")
        plt.bar(range(train[features].shape[1]), importances[indices],
               color="r", yerr=std[indices], align="center")
        plt.xticks(range(train[features].shape[1]), indices)
        plt.xlim([-1, train[features].shape[1]])
        plt.show()
    #########
```

```python
def print_results(prnt,pred_value, pred_proba, oob_score, acc_score, \
    class_report, conf_matrix, ras, mcc, cvs):
    #print all the stored results 
    if prnt==True:
        print('Predicted Values: \n',pred_value)
        print('OOB Score:\n',oob_score)
        print('Accuracy Score: \n',acc_score)
        print('Classification Report:\n',class_report)
        print('Confusion Matrix:\n', conf_matrix)
        print('ROC-AUC Score: \n', ras)
        print("Mathhews Correlation Coeff: \n",mcc)
        print("Mean Cross Val Score: \n", cvs)
```

## Methods for sampling

```python
def sample(df, direction,target_col='stable?', majority_label=0):
    """
    Takes a dataframe and either downsamples or up samples it. 
    Majority class is 0 by default.
    
    Code adapted from https://elitedatascience.com/imbalanced-classes
    """
    #split dataframe into majority and minority classes
    df_majority = df[df[target_col]==majority_label]
    df_minority = df[df[target_col]!=majority_label]
    
    #create resampled dataframe fragments
    if direction=='down':
        new_df = resample(df_majority, replace=False,n_samples=len(df_minority),random_state=0)
        other = df_minority    
    elif direction=='up':
        new_df = resample(df_minority, replace=True,n_samples=len(df_majority),random_state=0)
        other = df_majority
    else:
        print('Invalid sample direction')
    
    #concatenate fragments to resampled dataframe
    df_resampled = pd.concat([new_df,other])
    return df_resampled
    
    
```

```python
def stratify(df, library_list):
    new_df = pd.DataFrame()
    #for each library, stratify on 'stable?' and make a new df
    for i in library_list:
        try:
            new_df = new_df.append(sample(df[df['library']==i],'down'))
        except:
            new_df = new_df.append(sample(df[df['library']==i],'up'))
    return new_df
```

```python
def ks_analysis(df, actual_values_col, cutoff, stacked, predicted_values=None, target_col='predicted_values', features=['S_PC','Mean_H_entropy','Mean_L_entropy','Mean_E_entropy',\
                                'Mean_res_entropy','SumH_entropies','SumL_entropies','SumE_entropies'],density=True,split_all=True,topology_name='all'):
    """
    Outputs a KS dataframe and KS histograms with values that meet a certain cutoff parameter.
    
    Requires a df, actual_values_col and predicted_values parameter. 
    Entropy features are set as the default analytical target.
    
    TODO: Add a test for if predicted_values are None and no dataframe is being made
    """
    #adds a columns with predicted values on dataframe
    df['predicted_values'] = predicted_values
    
    #creates a df with the ks values
    final_df = ks_stat_df(df=df, avc=actual_values_col, cutoff=cutoff, target_col=target_col, features=features,split_all=split_all)
    
    #plots ks values
    histograms = ks_histograms(df, stacked=stacked, features=features,cutoff=cutoff, target_col=target_col,density=density,split_all=split_all,topology_name=topology_name)
        
    return final_df
```

```python
def ks_stat_df(df,avc,target_col,features,split_all,cutoff):     
    """
    Takes a dataframe that contains y_pred and
    y_actual values and creates a dataframe
    with its ks stat values and pvalues.
    `
    Must specify what the target_col is called (i.e,
    'pred_values', 'mr_class_predictions', etc)
    
    avc refers to actual_values_column (i.e. 'stable?')
    """
    #create True Positive and False Negative values
    actual_positive = df[df[avc]==1]
    tp = ("TP",actual_positive[actual_positive[target_col]==1])
    fn = ("FN",actual_positive[actual_positive[target_col]==0])
    

    #create True Negative and False Positive values
    actual_negative = df[df[avc]==0]
    tn = ("TN",actual_negative[actual_negative[target_col]==0])
    fp = ("FP",actual_negative[actual_negative[target_col]==1])
    
    #TODO: Write a check_notempty_df function for here
    
   # print(tp, fn, tn, fp)
    
    df_final = pd.DataFrame()
    if split_all:
        for pair in combinations([tp,fn,tn,fp],2):
            df1 = pair[0][1]
            df2 = pair[1][1]
            df = ks_df(df1,df2,features)
            df['name'] = [pair[0][0]+pair[1][0]]*len(df)
            df_final = df_final.append(df)

        df_final = df_final.set_index([df_final.index,df_final['name']])
        df_final.drop('name',axis=1,inplace=True)
        df_final = df_final[df_final.loc['KS stat'] > cutoff]
        df_final.dropna(axis=1, inplace=True)
    else:
        df = ks_df(actual_positive, actual_negative, features)
        df['name'] = ['PositiveNegative']*len(df)
        df_final = df_final.append(df)
        df_final.set_index([df_final.index,df_final['name']],inplace=True) #creates a multiindex 
        df_final = df_final[df_final.loc['KS stat'] > cutoff]
        df_final.dropna(axis=1,inplace=True)
    return df_final
```

```python
def ks_histograms(df, features, cutoff,density,split_all,avc='stable?',target_col='mr_class_predictions',stacked=False, topology_name='all'): 
    """
    Plots histograms of a given 
    KS value dataframe where the 
    KS values are above a given 
    cutoff
    """
    #create True Positive and False Negative values
    actual_positive = df[df[avc]==1]
    tp = ("TP",actual_positive[actual_positive[target_col]==1.0])
    fn = ("FN",actual_positive[actual_positive[target_col]==0.0])

    #create True Negative and False Positive values
    actual_negative = df[df[avc]==0]
    tn = ("TN",actual_negative[actual_negative[target_col]==0.0])
    fp = ("FP",actual_negative[actual_negative[target_col]==1.0])

    if split_all:
        #iterates through combinations of tuple list, plotting histograms
        plot_histograms([tp,tn,fp,fn],features,cutoff,density,split_all,stacked=stacked,topology_name=topology_name)
    else:
        plot_histograms([actual_positive,actual_negative], features, cutoff, density,split_all,topology_name=topology_name)
```

```python
def plot_histograms(list_of_tuples,features,cutoff,density,split_all,stacked=False,topology_name='all'):
    color_list = ['blue','pink','red','skyblue']
    if split_all:
        if stacked:
            tp = list_of_tuples[0][1]
            tn = list_of_tuples[1][1]
            fp = list_of_tuples[2][1]
            fn = list_of_tuples[3][1]
            for i in features:
                sns.kdeplot( tp[i], shade=True,color='blue',label='TP')
                sns.kdeplot( fp[i], shade=True,color='pink',label='FP')
                sns.kdeplot( tn[i], shade=True,color='red', label='TN')
                sns.kdeplot( fn[i], shade=True,color='skyblue', label='FN')
                plt.title('%s for %s topologies' %(i,topology_name))
                plt.ylabel('Probability')
                plt.xlabel('%s value' %i)
                plt.legend()
                plt.show()
            return
        for pair in combinations(list_of_tuples,2):
            df1 = pair[0][1]
            label1 = pair[0][0]
            df2 = pair[1][1]
            label2 = pair[1][0]
            ks_values = ks_df(df1,df2,features)
            signf_ks_values = ks_values[ks_values>cutoff]
            for column in signf_ks_values:
                if math.isnan(signf_ks_values[column][0]):
                    continue
                histogram([df1,df2],column, labels=[label1,label2],density=density)        
    else:
        df1 = list_of_tuples[0]
        label1 ='Positives'
        df2 = list_of_tuples[1]
        label2 = 'Negatives'
        ks_values = ks_df(df1,df2,features)
        signf_ks_values = ks_values[ks_values>cutoff]
        for column in signf_ks_values:
            if math.isnan(signf_ks_values[column][0]):
                    continue
            histogram([df1,df2],column, labels=[label1,label2],density=density)   
```

```python
def conf_matrix_values(value_tuple):
    """
    Takes tuple returned by random_forest() and 
    returns dataframes corresponding to the 
    true positive, true negative, false 
    positive and false negative values. 
    """
    df = value_tuple[0]
    train = value_tuple[1]
    test  = value_tuple[2]
    features = value_tuple[3]
    target = value_tuple[4]
    pred_values = value_tuple[5]
    
    test['pred_values'] = pred_values
    
    actual_stable = test[test['stable?']==1]
    tp = actual_stable[actual_stable['pred_values']==1]
    fn = actual_stable[actual_stable['pred_values']==0]
    

    actual_unstable = test[test['stable?']==0]
    fp = actual_unstable[actual_unstable['pred_values']==1]
    tn = actual_unstable[actual_unstable['pred_values']==0]
    
    return tp, tn, fp, fn
```

## Helper Functions

```python
def clean_columns(dataframe, columns):
    df = dataframe
    return df.drop(labels=columns, axis=1)
    
```

```python
def histogram(dataframe,attribute,labels,density): #takes a max of two dataframes
    if type(dataframe) == list:
        df1 = dataframe[0]
        df2 = dataframe[1]
        label1 = labels[0]
        label2 = labels[1]
        value = pd.to_numeric(df1[attribute].tolist())
        plt.hist(value,density=density,label=label1)
        value2  = pd.to_numeric(df2[attribute].tolist())
        plt.hist(value2, alpha=.75,density=density,label=label2)
        plt.legend()
        plt.title(attribute+" ("+label1+' , '+label2+')')
        plt.show()
    else:
        values = pd.to_numeric(dataframe[attribute].tolist())
        plt.hist(values, density=density, label=labels)
        plt.legend()
        plt.title(attribute+labels)
        plt.show()
  
```

```python
def ks_df(df1, df2, features):
    """
    Creates a dataframe comparing ks values for two 
    dataframes (df1, df2).
    """
    ks_df = pd.DataFrame()
    
    for column in features:
        ks_df[column] = ks_2samp(df1[column].values.tolist(),df2[column].values.tolist())

    ks_df['index'] = ['KS stat','pvalue']
    ks_df.set_index('index',inplace=True)
    #print('THIS IS KS DF:',ks_df)
    return ks_df
```

```python
def normalized_df(df,file_name):
    """
    This function takes a file_name that contains a list of names
    and uses them to construct a new dataframe that only contains
    the values of the specified names.
    
    >>> df = pd.read_csv('/home/jupyter/tacc-work/Jan/proteins.df/final_df.csv')
    >>> training_names = '/home/jupyter/tacc-work/Jan/train_names'
    >>> norm_df = normalized_df(df, training_names)
    >>> len(norm_df) < len(df)
    True
    """
    new_df = pd.DataFrame() #creates a new empty DataFrame
    file = open(file_name,'r')  #opens the file containing a list of names
    for name in file: #iterates through the names in the file
        try:
            name = name.rstrip() #strips the name of '\n'
            value = pd.DataFrame(df.loc[name])
            if value.index.contains(name): #returns boolean on whether it contains a specified key
                new_df = pd.concat([new_df,value])
                continue
            else:
                value = value.transpose()
                new_df = pd.concat([new_df,value])
                continue
            #print(new_df)
        except KeyError:
            continue
    return new_df
```

```python
def split_df(df):
    """
    Splits dataframe into training and 
    testing data using np.random.uniform
    instead of train_test_split. 
    """
    df['is_train'] = np.random.uniform(0,1,len(df)) <=.65 #separates the dataframe by making 65% have the positive 'is_train' label
    train, test = df[df['is_train']==True],df[df['is_train']==False] #separates into training and testing data
    df = df.drop(columns='is_train',axis=1) #drops the column 'is_train' from df
    train = train.drop(columns='is_train',axis=1) #drops the column 'is_train' from train df
    test = test.drop(columns='is_train',axis=1) #drops the column 'is_train' from test df
    return train,test
```

```python
def topology_si_values(entry, top_binary,max_value=True):
    values = []
    for i in range(len(entry[top_binary])):
        #print(h_bin[i],si_val[i])
        val = entry[top_binary][i]*entry['Si_values'][i]
        if val != 0:
            values.append(entry['Si_values'][i])
    if max_value:
        return max(values)
    return min(values)
        
```

```python
def to_float(string):
    """
    Takes a string that contains the format of a list within
    and removes characters that impede the float function
    to work
    
    >>> df = pd.read_csv('/home/jupyter/tacc-work/Jan/proteins.df/rosettafinal_df.csv')
    >>> type(df['Si_values'][0])
    <class 'str'>
    
    >>> type(to_float(df['Si_values'][0]))
    <class 'list'>
    
    """
    try:
        string = string[1:-1]
        string = string.replace("'","")
        string = string.replace(",","")
        lst = string.split()
        return [float(i) for i in lst]
    except TypeError:
        print('string is:',string)
```
