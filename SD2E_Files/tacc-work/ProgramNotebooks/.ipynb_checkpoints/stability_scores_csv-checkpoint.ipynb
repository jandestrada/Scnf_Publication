```python
def canonicalize_name(name):
    if name.endswith('.seq'):
      i = name.rfind('.pdb')
      return name[0:i] + '.pdb'
    if not name.endswith('.pdb'):
      return name + '.pdb'
    else:
      return name

```

```python
#assumes you are at correct directory
def stability_scores(stability_file):
    stability_dict = {}
    #makes sure you are at right directory
    path_name=''
    if os.getcwd != '/home/jupyter/tacc-work/Jan/consistent_data/':
        os.chdir('/home/jupyter/tacc-work/Jan/consistent_data/')
        path_name = os.path.join(os.getcwd(),stability_file)
    else:
        path_name = os.path.join(os.getcwd(),stability_file)
    
    #finds where each term is in file and assigns an index value to each
    reader = csv.reader(open(stability_file))
    file_list = list(reader)
    header = file_list[0]
    name_index = int(header.index('name'))
    library_index = int(header.index('library'))
    ss_t_index = int(header.index('stabilityscore_t'))
    ss_c_index = int(header.index('stabilityscore_c'))
    ss_index = (header.index('stabilityscore'))
    total_score_index = int(header.index('total_score'))
    nres_index = int(header.index('nres'))
    nres_helix_index = int(header.index('nres_helix'))
    nres_loop_index = int(header.index('nres_loop'))
    nres_sheet_index = int(header.index('nres_sheet'))
    stable_index = int(header.index('stable?'))
    
    
    reader = csv.reader(open(stability_file))

    for line in reader:
        #define variables for name and stability scores
        if len(line) < 121:
            break
        pdb_name = canonicalize_name(line[name_index])
        splitext_name = os.path.splitext(pdb_name)
        name = splitext_name[0]
        library = line[library_index]
        stabilityscore_t = line[ss_t_index]
        stabilityscore_c = line[ss_c_index]
        stabilityscore = line[ss_index]
        nres = line[nres_index]
        nres_helix = line[nres_helix_index]
        nres_loop = line[nres_loop_index]
        nres_sheet = line[nres_sheet_index]
        total_score = line[total_score_index]
        stable = line[stable_index]
        #update dictionary with variables
        stability_dict[name]={
            'stabilityscore_t':stabilityscore_t,
            'stabilityscore_c':stabilityscore_c,
            'stabilityscore':stabilityscore,
            'nres':nres,
            'nres_helix':nres_helix,
            'nres_loop':nres_loop,
            'nres_sheet':nres_sheet,
            'total_score':total_score,
            'library':library,
            'stable':stable
        }
    return stability_dict
```

```python
df = pd.DataFrame((stability_scores(Hamed)))
df = df.drop('name',1)
df
```
