### Q: How imbalanced is the entire dataset?

```python
df = pd.read_csv("./dataframes/df_aggregated_sd2e.csv")
```

```python
df['stable?'] = df['stabilityscore']>1
```

```python
vals,counts = np.unique(df['stable?'],return_counts=True)
```

```python
total = counts[0] + counts[1]
percent_false = counts[0]/total
percent_true = counts[1]/total
```

### Answer:

```python
#results for 'stabilityscore'
print("percent false: %.2f\npercent true: %.2f"%(percent_false,percent_true))
```

```python
#results for 'stabilityscore_calibrated'

df['stable?_calibrated'] = df['stabilityscore_calibrated']>1

vals,counts = np.unique(df['stable?_calibrated'],return_counts=True)

total = counts[0] + counts[1]
percent_false = counts[0]/total
percent_true = counts[1]/total

print("percent false: %.2f\npercent true: %.2f"%(percent_false,percent_true))
```

# What does the sklearn Mutual information function say about Rosetta and Entropy features

```python
from sklearn.feature_selection import mutual_info_classif
```

```python
#load the data
df = pd.read_csv('./dataframes/df_aggregated_spc.csv')
df['stabilityscore_cnn_calibrated_2classes'] = df['stabilityscore_cnn_calibrated'] > 1
spc_features = pd.read_csv('./entropy_features.csv')["0"].to_list()
ros_features = pd.read_csv('./rosetta_features.csv')["0"].to_list()
```

```python
#features 
X = df[spc_features+ros_features]
y = df['stabilityscore_cnn_calibrated_2classes'].values
```

```python
mutual_info_classif(X,y)
```
