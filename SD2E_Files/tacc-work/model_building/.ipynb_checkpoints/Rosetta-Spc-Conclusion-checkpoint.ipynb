```python
def add_values(df, test_harness_scores):
    clf_scores_names = ['Accuracy',  'Balanced Accuracy',  'AUC Score',  'Average Precision',  'F1 Score', 'Precision',  'Recall']
    rgr_scores_names = ['R-Squared','RMSE']
    if len(test_harness_scores)==len(clf_scores_names):
        for i in range(len(clf_scores_names)):
            df[clf_scores_names[i]] = test_harness_scores[i]
    elif len(test_harness_scores)==len(rgr_scores_names):
        for i in range(len(rgr_scores_names)):
            df[rgr_scores_names[i]] = test_harness_scores[i]
    else: 
        print("Inputted scores do not match up length of values.")
            
    return df
```

```python
def make_multiindex_df(list_of_fragments, names_of_columns): 
    assert len(list_of_fragments)==len(names_of_columns), 'given list of columns does not match len of fragments'
    df = pd.DataFrame()
    original_columns = list(list_of_fragments[0].columns)
    for i in range(len(list_of_fragments)):
        frag = list_of_fragments[i].copy()
        frag.columns = pd.MultiIndex.from_product([[names_of_columns[i]],original_columns])
        
        df = pd.concat([df,frag],axis=1)
    return df
        
        
        
```

```python
def cross_val_plots(df,show_RS=True,show_R=True,show_S=True):
    for i in range(len(clf_scores_names)):
        score_rs = df.iloc[0,list(range(i,70,7))].values
        score_r = df.iloc[1,list(range(i,70,7))].values
        score_s = df.iloc[2,list(range(i,70,7))].values
        
        if show_RS:
            sns.lineplot(x=range(len(score_rs)),y=score_rs,label='R+S')
        if show_R:
            sns.lineplot(x=range(len(score_rs)),y=score_r,label='R')
        if show_S:
            sns.lineplot(x=range(len(score_rs)),y=score_s,label='S')
        
        #plt.legend(loc='center right')
        plt.title("%s per run"%clf_scores_names[i])
        plt.xlabel("Run number")
        plt.ylabel("%s"%clf_scores_names[i])
        plt.show()
        plt.close('all')
```

```python
pearson_list = []
names_list_x = []
names_list_y = []
for col in combinations(df_aggregated[ros_features+spc_features].columns,2):
    x = col[0]
    y = col[1]
    #print(x,y)
    
    pearson_list.append((pearsonr(df_aggregated[x],df_aggregated[y])))
    names_list_x.append(x)
    names_list_y.append(y)
```

```python
df_pearson = pd.DataFrame(pearson_list,columns=['Pearson coeff.','p-value'])
df_pearson['Feature_x'] = names_list_x
df_pearson['Feature_y'] = names_list_y

df_pearson['is_rosetta_x'] = df_pearson['Feature_x'].isin(ros_features)
df_pearson['is_rosetta_y'] = df_pearson['Feature_y'].isin(ros_features)
```

```python
df_pearson_x_ros = df_pearson[df_pearson['is_rosetta_x']==True]
df_pearson_x_ros_y_spc = df_pearson_x_ros[df_pearson_x_ros['is_rosetta_y']==False]
```
