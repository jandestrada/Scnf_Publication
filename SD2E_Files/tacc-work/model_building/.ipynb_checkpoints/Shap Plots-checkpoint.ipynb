```python
import pandas as pd
import numpy as np
import random
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import export_graphviz
from math import sqrt, fabs
import os
import matplotlib.pyplot as plt
from operator import itemgetter
import shap
```

```python
def dummy_df():
    df = pd.DataFrame()
    df['feature1'] = np.arange(1,2001)
    df['feature2'] = df['feature1'] * 2
    df['feature3'] = df['feature2'] * 3
    df['feature4'] = [random.randint(10,20) for i in range(2000)]
    df['feature5'] = [random.randint(10,20)]*2000
    df['target'] = [0]*1000 + [1]*1000
    return df
```

```python
df = dummy_df()
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,:5],df.iloc[:,5])

clf = DecisionTreeClassifier()
clf.fit(X_train,y_train)
```

```python
def perform_shap_audit(df,clf,X_train,X_test):
        """
        From:
        https://slundberg.github.io/shap/notebooks/Census%20income%20classification%20with%20scikit-learn.html
        """
        features = df.columns[:5]
        classifier = clf
        train_X = X_train
        test_X = X_test

        shap.initjs()
        f = lambda x: classifier.predict_proba(x)[:,1]
        train_X_df = pd.DataFrame(data = train_X, columns = features)
        med = train_X_df.median().values.reshape((1,train_X_df.shape[1]))
        
        #print(train_X_df)
        #print(train_X_df.shape[1])
        #print(med)
        
        explainer = shap.KernelExplainer(f, med)
        test_X_df = pd.DataFrame(data = test_X, columns = features)
        shap_values = explainer.shap_values(test_X_df)
        means = []
        totals_list = [0.0 for f in features]
        for val_list in shap_values:
            for i in range(0, len(val_list)):
                totals_list[i] += fabs(val_list[i])
        means = [(feat, total / len(shap_values)) for feat, total in zip(features, totals_list)]
        mean_shaps = sorted(means, key=itemgetter(1), reverse=True)
        print(mean_shaps)   
        
        
        #plots!
        output_path = "/home/jupyter/tacc-work/model_building/dataframes/saved_plots/shap/shap_native_plots/"
        
        
        #This plot shows no data, but it appears!
        shap.force_plot(explainer.expected_value, shap_values[0,:], test_X.iloc[0,:],matplotlib=True,show=True)
        plt.savefig(os.path.join(output_path,"shap_test_fig1.jpeg"),bbow_inches="tight")
        plt.close()
        
        #I cannot see this plot for some reason
        shap.force_plot(explainer.expected_value, shap_values, train_X,show=True)
        plt.savefig(os.path.join(output_path,"shap_test_fig2.jpeg"),bbow_inches="tight")
        plt.close()
        
        #dependencies plot
        shap.dependence_plot("feature1", shap_values,test_X,interaction_index='feature3',show=True)
        plt.savefig(os.path.join(output_path,"shap_test_fig3.jpeg"),bbow_inches="tight")
        plt.close()
        
        #Summary of effect of feature on predictions
        shap.summary_plot(shap_values, test_X,show=True)
        plt.savefig(os.path.join(output_path,"shap_test_fig3a.jpeg"),bbow_inches="tight")
        plt.close()

        shap.summary_plot(shap_values, test_X, plot_type="bar",show=True)
        plt.savefig(os.path.join(output_path,"shap_test_fig3b.jpeg"),bbow_inches="tight")
        plt.close()
        
         
        
        
        #return mean_shaps
```

```python
import warnings
warnings.filterwarnings('ignore')

perform_shap_audit(df,clf,X_train,X_test)
```
